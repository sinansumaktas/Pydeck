```python
import pydeck as pdk
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\sinan.sumaktas\Desktop\Veriler\park_location.csv")
```

```python
layer = pdk.Layer(
    "HexagonLayer",
    df,
    get_position=['LONGITUDE', 'LATITUDE'],
    auto_highlight=True,
    elevation_scale=50,
    pickable=True,
    elevation_range=[0, 3000],
    extruded=True,
    coverage=1)
```

```python
view_state = pdk.ViewState(
    longitude=33,
    latitude=36,
    zoom=6,
    min_zoom=5,
    max_zoom=15,
    pitch=40.5,
    bearing=-27.36)
r = pdk.Deck(layers=[layer], initial_view_state=view_state)
r.to_html('demo.html', notebook_display=False)
r
```
