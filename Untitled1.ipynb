```python
import pydeck as pdk
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\sinan.sumaktas\Desktop\Veriler\salk-kurum-ve-kurulularna-ait-bilgiler.csv")
```

```python
layer = pdk.Layer(
    "ScatterplotLayer",
    df,
    pickable=True,
    opacity=0.8,
    stroked=True,
    filled=True,
    radius_scale=6,
    radius_min_pixels=1,
    radius_max_pixels=100,
    line_width_min_pixels=1,
    get_position=['BOYLAM','ENLEM'],
    get_radius="exits_radius",
    get_fill_color=[255, 140, 0],
    get_line_color=[0, 0, 0],
)


view_state = pdk.ViewState(latitude=36, longitude=33, zoom=10, bearing=0, pitch=0)


r = pdk.Deck(layers=[layer], initial_view_state=view_state, tooltip={"text": "{ADI}\n{ILCE_ADI}"})
r.to_html("scatterplot_layer.html")
```
